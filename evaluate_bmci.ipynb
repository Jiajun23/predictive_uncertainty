```python
labels = ["0.1", "0.3", "0.5", "0.7", "0.9"]

def interval_errors(model, y, cwv):
    plt.figure()
    ratios = np.zeros((cwv.shape[0], 5))
    bins   = np.linspace(0,1,101)
    for i in range(5):
        t_l = intervals[:,4-i]
        t_r = intervals[:,6+i]
        hits = (cwv > t_l.reshape(-1,1)) * (cwv < t_r.reshape(-1,1))
        ratios[:,i] = (hits.sum(axis=1, keepdims=True) / hits.shape[1]).ravel()
        plt.hist(ratios[:,i], bins=bins, normed=True, label=labels[i])
    return ratios

errs = interval_errors(model, v_y_true, v_cwv)
plt.legend()
np.mean(errs, axis=0)
```
